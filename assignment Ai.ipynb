```python
#rabbit problem using bfs
from collections import deque


initial_state = ['E', 'E', 'E', '_', 'W', 'W', 'W']
goal_state = ['W', 'W', 'W', '_', 'E', 'E', 'E']

def get_next_states(state):
    states = []
    index = state.index('_')
    directions = [-1, 1, -2, 2]

    for d in directions:
        new_index = index + d
        if 0 <= new_index < len(state):
           
            if d == -1 and state[new_index] == 'E':
              
                new_state = state[:]
                new_state[index], new_state[new_index] = new_state[new_index], new_state[index]
                states.append(new_state)
            elif d == 1 and state[new_index] == 'W':
               
                new_state = state[:]
                new_state[index], new_state[new_index] = new_state[new_index], new_state[index]
                states.append(new_state)
            elif d == -2 and state[index - 1] == 'W' and state[new_index] == 'E':
               
                new_state = state[:]
                new_state[index], new_state[new_index] = new_state[new_index], new_state[index]
                states.append(new_state)
            elif d == 2 and state[index + 1] == 'E' and state[new_index] == 'W':
               
                new_state = state[:]
                new_state[index], new_state[new_index] = new_state[new_index], new_state[index]
                states.append(new_state)

    return states

def bfs(start, goal):
    queue = deque([[start]])
    visited = set()

    while queue:
        path = queue.popleft()
        state = path[-1]
        state_tuple = tuple(state)
        
        if state == goal:
            return path

        if state_tuple in visited:
            continue

        visited.add(state_tuple)

        for next_state in get_next_states(state):
            queue.append(path + [next_state])

    return None


solution = bfs(initial_state, goal_state)


if solution:
    print("Solution found in", len(solution) - 1, "moves:\n")
    for step in solution:
        print(step)
else:
    print("No solution found.")

#rabbit problem using dfs
def is_goal(state):
    return state == "WWW_EEE"

def get_next_states(state):
    states = []
    pos = state.index('_')
    state_list = list(state)
    
   
    if pos > 0 and state[pos - 1] == 'E':  
        new_state = state_list[:]
        new_state[pos], new_state[pos - 1] = new_state[pos - 1], '_'
        states.append("".join(new_state))
    
    if pos > 1 and state[pos - 2] == 'E' and state[pos - 1] == 'W': 
        new_state = state_list[:]
        new_state[pos], new_state[pos - 2] = new_state[pos - 2], '_'
        states.append("".join(new_state))
    
   
    if pos < len(state) - 1 and state[pos + 1] == 'W':  
        new_state = state_list[:]
        new_state[pos], new_state[pos + 1] = new_state[pos + 1], '_'
        states.append("".join(new_state))
    
    if pos < len(state) - 2 and state[pos + 2] == 'W' and state[pos + 1] == 'E':  
        new_state = state_list[:]
        new_state[pos], new_state[pos + 2] = new_state[pos + 2], '_'
        states.append("".join(new_state))
    
    return states

def dfs(state, path, visited):
    if is_goal(state):
        return path + [state]
    
    for next_state in get_next_states(state):
        if next_state not in visited:
            visited.add(next_state)
            result = dfs(next_state, path + [state], visited)
            if result:
                return result
    return None

if __name__ == "__main__":
    initial_state = "EEE_WWW"
    solution = dfs(initial_state, [], {initial_state})
    if solution:
        print("Solution Path:")
        for step in solution:
            print(step)
    else:
        print("No solution found.")
```
